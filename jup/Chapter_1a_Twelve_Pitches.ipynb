## Musical Content in Binary

The Western harmonic system begins with twelve notes. These twelve notes can be written as single numbers of twelve bits, as follows:

```python
all_notes = 0b111111111111

import jupyter_pt_tools as jpt_pt
import IPython.display as ipd

jpt_pt.notegroup_heatmap(all_notes, chromatic=True, title="Chromatic Scale")
ipd.Audio(jpt_pt.notegroup_wavestep(0b111111111111, chromatic=True), rate=4000)
```

This type of organization of pitches in a set of twelve creates what is (somewhat) commonly known as a 'pitch-class set'. 

It is a conventional and powerful system for music analysis. We will reconfigure in these chapters to expose a wealth of in-built harmonic material.

It is *crucial* that the pitch-class set be represented as an *integer*, because this analysis system relies **almost entirely on logic operations**. This gives it the most profound efficiency that a computer can offer, from the very basement of its operations.

---

We can also view them in the order of the 'circle of fifths. It's the same notes, but not viewed in order low-to-high, but in clarity of ratio: each note's frequency is (approximately) in a ratio of 3:2 (if it goes up) or 3:4 (if it goes down) compared to the note before it.

```python
jpt_pt.notegroup_heatmap(all_notes, chromatic=False, title="Chromatic Scale in Circle of Fifths Order")
chromatic_order = [0b100000000000 >> i for i in range(12)]
ipd.Audio(jpt_pt.link_wavestep_sequences(chromatic_order), rate=4000)
```

## Organized Subsets of Pitches 1 -- A C Major Scale

Twentieth-century theory notwithstanding, it remains rare to hear all twelve notes outside of ornaments, slides, and special effects. The most conventional musical settings use a *scale* -- a subset of the twelve notes, usually between five and eight notes. The most common in Western music is the major scale.

```python
# and then we can make an image (or two)

jpt_pt.notegroup_heatmap(0b101011010101, chromatic=True)
ipd.Audio(jpt_pt.notegroup_wavestep(0b101011010101, chromatic=True), rate=4000)
```

The pattern above is the familiar piano keyboard: black keys, white keys.

## Organized Subsets of Pitches 2 -- A C Major Chord

Last, if there aren't too many notes (usually five or less), and they are well spaced (more about that later), we can hear them as chords:

```python
c_major = 0b100010010000
jpt_pt.notegroup_heatmap(c_major, chromatic=True)

c_major_signal = jpt_pt.notegroup_wavepile(c_major, chromatic=True)
ipd.Audio(c_major_signal, rate=4000)
```

These are some simple musical elements, represented in binary.
