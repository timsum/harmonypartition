## The Harmonic Identity Matrix

The first seven of 12 bits can be considered as the harmonic identity matrix. Simple though it looks, it carries within it an enormous amount of information. 
Its unique subdivisions define meaningful harmonic possibilities.

```python
c_major = 0b111111100000
jpt_pt.notegroup_heatmap(c_major)
ipd.Audio(jpt_pt.notegroup_wavestep(c_major), rate=4000)
```

## Scales

Consider, first, how to build a scale in this context. As discussed before, we can pick a note and follow **every second note** thereafter to form a scale.

```python
seq_length = 7
scale_seq = [np.array([0, 0, 1, 2, i]) for i in range(seq_length)]
bin_seq = [pt_keypattern.get_binary_KPDVE_chord(a_kpdve) for a_kpdve in scale_seq]
yticks = [pt_naming_conventions.chord_function_in_key(a_kpdve) for a_kpdve in scale_seq]
# draw binary sequences
jpt_pt.multiple_notegroup_heatmap(bin_seq, yticks=yticks)
# listen to piling-up scale
patterns = [mode + disp for mode, disp in [pt_musicutils.get_chord_disp_tuple(kpdve) for kpdve in scale_seq]]
sequence = jpt_pt.link_ordered_wavepile_sequences(patterns)
ipd.Audio(sequence, rate=4000)
```

# Chord Sequences

Here is a sample chord sequence: roughly, 'I've got rhythm': I, vi, ii, V.

You have probably noticed the string KPDVE in the previous examples -- it is, first, an efficient system for creating these binary sequences, but more than that, it will prove undergird a system of music and audio analysis with a strong relationship to patterns of machine learning.

```python
seq_length = 4

ground_seq = [np.array([0, 0, seq_length-i,4,3]) for i in range(seq_length)]
kpdve_seq = [ground_seq[3]]
kpdve_seq += ground_seq
kpdve_seq += ground_seq
kpdve_seq += ground_seq[:3]
kpdve_seq += ground_seq[1:]

bin_seq = [pt_keypattern.get_binary_KPDVE_chord(a_kpdve) for a_kpdve in kpdve_seq]
yticks = [pt_naming_conventions.chord_function_in_key(a_kpdve) for a_kpdve in kpdve_seq]

jpt_pt.multiple_notegroup_heatmap(bin_seq, yticks=yticks)

sequence = jpt_pt.link_wavepile_sequences(bin_seq, duration=10)
ipd.Audio(sequence, rate=4000)
```
