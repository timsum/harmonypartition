The preceding chapters have introduced some tools for conceiving and hearing bit-level information as musical pattern.

This chapter will organize bit-level organization on musical terms, according to five integers.

```python
import numpy as np

import IPython.display as ipd

import pt_keypattern
import pt_naming_conventions
import pt_musicutils

import jupyter_pt_tools as jpt_pt
```

# Five Variables

Hidden layers:
* K "Key" (0-11) in the circle of fifths
* P "Pattern" (0-7) as a distortion of a central key (relative minor, harmonic minor, harmonic major)

Audible layers:
* D "Degree" (0-7) as the distance of a chord from the Tonic chord
* V "Voicing" (0-7) as the regular interval ('stride') between notes in a group
* E "Extensions" (0-7) as the distance in notes from the root of a chord to the furthest (measured in increments of voicing "V")

These five variables will control the sounds on this page, and we can introduce them one by one. The starting point is below. It will yield an F major chord, which you can see and hear using the tools from previous chapters.

```python
k = 0 # C Major
p = 0 # no distortions for minor
d = 1 # Chord root on C
v = 4 # Use every fourth note
e = 6 # Use all the 7 notes (0-6 inclusive)

page_kpdve = np.array([k,p,d,v,e])
```

***

## K (Key)

```python
# ==========================================================================================================
k = 3

# change this variable (between 0-11) to see how change the central key of the scale (then shift-enter)
# ==========================================================================================================

page_kpdve = np.array([k,0,1,4,6])
page_chord = pt_keypattern.get_binary_KPDVE_chord(page_kpdve)
k_description = pt_naming_conventions.circle_fifth_notes(k=k)[(k+1)%12]+ " Major"

mode, disp = pt_musicutils.get_tonic_mode_disp_tuple(page_kpdve)
notes = mode + disp

jpt_pt.notegroup_heatmap(page_chord, title=k_description)
ipd.Audio(jpt_pt.ordered_notegroup_wavestep(notes), rate=4000)
```

## P (Pattern)

```python
# ==========================================================================================================
p = 5 

# change this variable (between 0-6) to see how change the central key of the scale (then shift-enter)
# ==========================================================================================================

page_kpdve = np.array([k, p, 1, 4, 6])
page_chord = pt_keypattern.get_binary_KPDVE_chord(page_kpdve)
p_description = pt_naming_conventions.conv_tonic_name_for_kpdve(page_kpdve) + " " + pt_naming_conventions.PATTERN_CONVENTIONAL_NAMES[p]

mode, disp = pt_musicutils.get_tonic_mode_disp_tuple(page_kpdve)
notes = mode + disp

jpt_pt.multiple_notegroup_heatmap([page_chord], title=p_description)
ipd.Audio(jpt_pt.ordered_notegroup_wavestep(notes), rate=4000)
```

## D (Degree)

```python
# ==========================================================================================================
d = 1 

# change the 'd' variable (between 0-6) to see how change the central key of the scale (then shift-enter)
# ==========================================================================================================
page_kpdve = np.array([k, p, d, 4, 2])
page_chord = pt_keypattern.get_binary_KPDVE_chord(page_kpdve)
function = pt_naming_conventions.chord_function_in_key(page_kpdve)
notes = pt_naming_conventions.chord_note_names_for_KPDVE(page_kpdve)
d_description = "Root and Function: " + function + " (in " + p_description + ") "  + str(notes)

jpt_pt.multiple_notegroup_heatmap([page_chord], title=d_description)
ipd.Audio(jpt_pt.notegroup_wavepile(page_chord), rate=4000)
```

## V (Voicing)

```python
# ==========================================================================================================
v = 4 

# change this variable (between 0-6) to see how change the central key of the scale (then shift-enter)
# ==========================================================================================================

page_kpdve = np.array([k, p, d, v, 3])
page_chord = pt_keypattern.get_binary_KPDVE_chord(page_kpdve)
notes = pt_naming_conventions.chord_note_names_for_KPDVE(page_kpdve)

v_description = "Voicing: "  + pt_naming_conventions.VOICING_NAMES[v] + " "  + str(notes)

jpt_pt.multiple_notegroup_heatmap([page_chord], title=v_description)
ipd.Audio(jpt_pt.notegroup_wavepile(page_chord), rate=4000)
```

## E (Extensions)

```python
# ==========================================================================================================
e = 2 

# change this variable (between 0-6) to see how change the central key of the scale (then shift-enter)
# ==========================================================================================================

page_kpdve = np.array([k, p, d, v, e])
page_chord = pt_keypattern.get_binary_KPDVE_chord(page_kpdve)
notes = pt_naming_conventions.chord_note_names_for_KPDVE(page_kpdve)

e_description = "Notes in Chord: " + pt_naming_conventions.EXTENSION_NAMES[e] + " "  + str(notes)

jpt_pt.multiple_notegroup_heatmap([page_chord], title=e_description)
ipd.Audio(jpt_pt.notegroup_wavepile(page_chord), rate=4000)
```
