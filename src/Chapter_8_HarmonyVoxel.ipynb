```python
%matplotlib notebook
#%matplotlib widget
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

import pt_utils
import pt_keypattern
import pt_naming_conventions

from mpl_toolkits.mplot3d import axes3d
```

```python
def get_colors_for_shape(a_shape):
    x, y, z = np.indices(a_shape)
    vox1 = (x == 0)
    vox2 = (x == 1)
    vox3 = (x == 2)
    vox4 = (x == 3)
    vox5 = (x == 4)
    vox6 = (x == 5)
    vox7 = (x == 6)
    vox8 = (x == 7)
    vox9 = (x == 8)
    vox10 = (x == 9)
    vox11 = (x == 10)
    vox12 = (x == 11)

    voxels = vox1 | vox2 | vox3 | vox4 | vox5 | vox6 | vox7 | vox8 | vox9 | vox10 | vox11 | vox12

    colors = np.empty(voxels.shape, dtype=object)
    colors[vox1] = 'red'
    colors[vox2] = 'orangered'
    colors[vox3] = 'orange'
    colors[vox4] = 'gold'
    colors[vox5] = 'yellow'
    colors[vox6] = 'yellowgreen'
    colors[vox7] = 'green'
    colors[vox8] = 'lightseagreen'
    colors[vox9] = 'blue'
    colors[vox10] = 'mediumslateblue'
    colors[vox11] = 'purple'
    colors[vox12] = 'mediumvioletred'
    
    return colors
```

```python
int_list = np.array([[pt_keypattern.get_binary_KP(k, p) for p in range(1)] for k in range(12)])
array_list = np.array([[pt_utils.binary_notegroup_to_numpy_array(p) for p in int_list[k]] for k in range(12)])

full_shape = np.zeros((12,12,12))
full_shape[:,0] = array_list.squeeze()

colors = get_colors_for_shape(full_shape.shape)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.voxels(full_shape, edgecolors='w', facecolors=colors, shade=True)
plt.show()
```

```python
# make a map of a key, with its patterns as X, its degrees as y
# meshgrid is p and d: the surface values are the binary values numbers of complete scales
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x =  np.arange(0, 12, 1)
y = np.arange(0, 7, 1)
X, Y = np.meshgrid(x, y)

int_list = np.array([[pt_keypattern.get_binary_KPDVE_chord(np.array([k, 0, p, 4, 6])) for p in range(7)] for k in range(12)])
zs = np.array(np.log(int_list))
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('keys')
ax.set_ylabel('degrees')
ax.set_zlabel('values')

plt.show()
```
